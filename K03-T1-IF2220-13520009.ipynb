Menentukan setiap kolom numerik berdistribusi normal atau tidak dari histogram sebelumnya.

```python
def normalTest(df, i):
  alpha = 0.05
  val, p_value = st.normaltest(df[df.columns[i]])
  print("P-value: ", p_value)
  if(p_value>alpha):
    print(df.columns[i] + " berdistribusi normal")
  else:
    print(df.columns[i] + " tidak berdistribusi normal")

for i in range(11):
  normalTest(df, i)
```

### a. Nilai Rata-rata pH di atas 7?

```python
phbp = df.boxplot(["pH"])
phbp.set_title("Boxplot pH")
plt.show()
```

1. $H_0: \mu_{pH} = 7 $
2. $H_1: \mu_{pH} \gt 7 $
3. $\alpha = 0.05$
4. Uji statistik dengan $ z = \frac{mean-\mu}{\frac{\sigma}{\sqrt{n}}}$ dan daerah kritis $z > z_{\alpha} = 1.6448$
5. Menghitung nilai uji statistik

```python
mean = 7
mu = df["pH"].mean()
std = df["pH"].std()
n = df["pH"].count()
z = (mu - mean) * (n**(0.5)) / std
p_val = 1 - st.norm.cdf(abs(z))
print("z = " + str(z))
print("p-value = " + str(p_val))
```

6. Karena nilai $z = 2.485 > z_a$ sehingga berada pada daerah kritis, maka hipotesis $H_0$ ditolak dan rata-rata pH lebih dari 7

```python
hdbp = df.boxplot(["Hardness"])
hdbp.set_title("Boxplot Hardness")
plt.show()
```

1. $H_0: \mu_{Hardness} = 205 $
2. $H_1: \mu_{Hardness} \neq 205 $
3. $\alpha = 0.05$
4. Uji statistik dengan $ z = \frac{mean-\mu}{\frac{\sigma}{\sqrt{n}}}$ dan daerah kritis $z < -z_{\frac{\alpha}{2}}$ atau $z > z_{\frac{\alpha}{2}}$ dengan $z_{\frac{\alpha}{2}} = 1.9599$
5. Menghitung nilai uji statistik

```python
mean = 205
mu = df["Hardness"].mean()
std = df["Hardness"].std()
n = df["Hardness"].count()
z = (mu - mean) * (n**(0.5)) / std
p_val = (1 - st.norm.cdf(abs(z)))*2
print("z = " + str(z))
print("p-value = " + str(p_val))
```

### d. Proporsi nilai Conductivity yang lebih dari 450, adalah tidak sama dengan 10%?

```python
cbp = df.boxplot(["Conductivity"])
cbp.set_title("Boxplot Conductivity")
plt.show()
```

1. $H_0: p = 0.1 $
2. $H_1: p \neq 0.1 $
3. $\alpha = 0.05$
4. Uji statistik dengan $ z = \frac{x-n*p_0}{\sqrt{n*p_0*q_0}}$ dan daerah kritis $z < -z_{\frac{\alpha}{2}}$ atau $z > z_{\frac{\alpha}{2}}$ dengan $z_{\frac{\alpha}{2}} = 1.9599$
5. Menghitung nilai uji statistik

```python
a = 0.05
p0 = 0.1
q0 = 1 - p0
x = 0
n = df["Conductivity"].count()
for i in range(len(df["Conductivity"])):
  if(df["Conductivity"][i] > 450):
    x+=1
z = (x - n*p0)/((n*p0*q0)**0.5)
p_val = (1 - st.norm.cdf(abs(z)))*2
print("z = " + str(z))
print("p-value = " + str(p_val))
```

6. Karena $z = 40.4463 > z_{\frac{\alpha}{2}}$ sehingga berada pada daerah kritis, maka hipotesis $H_0$ ditolak dan proporsi nilai Conductivity yang lebih dari 450 tidak sama dengan 10%

2. H1: $\mu_{A1} \neq \mu_{A2} $
3. $\alpha = 0.05$
4. Menentukan daerah kritis dengan distribusi normal dan uji statistik two-sample: test on two mean with known variances. note: Saat n>30, distribusi t tidak berbeda signifikan dengan distribusi normal dan S estimasi yang baik untuk $\sigma$. 

```python
za = st.norm.ppf(1-0.025)
print("Batas bawah: ",za)
print("Batas atas: ", -za)
```

Didapat daerah kritis: $z < -z_{\alpha/2}$ or $z > z_{\alpha/2}$: $z < -1.960$ or $z > 1.960$
5. Menghitung nilai uji statistik dan p-value

```python
from math import sqrt
n1 = 1005
n2 = 1005
x1 = df.loc[:1004, "Sulfate"].mean()
x2 = df.loc[1005:, "Sulfate"].mean()
s1_2 = df.loc[:1004, "Sulfate"].var()
s2_2 = df.loc[1005:, "Sulfate"].var()
z_value = (x1-x2)/sqrt(s1_2/n1 + s2_2/n2)
p_value = 2*min(1 - st.norm.cdf(abs(z_value)),st.norm.cdf(abs(z_value)))
print("Nilai z: ", z_value)
print("Nilai p: ", p_value)
```

2. $H1: P(X1>4) > p(X2>4)$
3. $\alpha = 0.05$
4. Menentukan daerah kritis dengan distribusi normal.

```python
z = st.norm.ppf(1-0.05)
print("nilai z-alpha = ", z)
```

5. Menghitung nilai uji statistik dan p-value

```python
n1 = 1005
n2 = 1005
awal = df.loc[:1004, "Turbidity"]
akhir = df.loc[1005:, "Turbidity"]
x1 = 0
x2 = 0
for val in awal:
  if(val>4):
    x1 +=1
for val in akhir:
  if(val>4):
    x2 +=1
p1 = x1/n1
p2 = x2/n2
p = (x1+x2)/(n1+n2)
z_value = (p1-p2) / sqrt(p*(1-p)*(1/n1 + 1/n2))
p_value = 1 - st.norm.cdf(abs(z_value))
print("nilai z: ", z_value)
print("nilai p: ", p_value)
```

```python
from math import sqrt
s1_2 = df.loc[:1004, "Sulfate"].var()
s2_2 = df.loc[1005:, "Sulfate"].var()
f_value = s1_2/s2_2
p_value = 2*min(1 - st.f.cdf(f_value, n1-1,n2-1),st.f.cdf(f_value, n1-1,n2-1))
print("Nilai f: ", f_value)
print("Nilai p: ", p_value)
```

6. Didapat nilai f di luar daerah kritis (0.884 < 1.015 < 1.13) dan p-value > $\alpha$. Oleh karena itu, keputusan yang diambil adalah tidak menolak H0 dan menyimpulkan variansi bagian awal kolom Sulfate tidak berbeda signifikan dengan variansi bagian akhir kolom Sulfate.

Test korelasi: tentukan apakah setiap kolom non-target berkorelasi dengan kolom target, dengan menggambarkan juga scatter plot nya. Gunakan correlation test.

```python
def testKolerasi(df, name):
  sns.regplot(df[name], df["Potability"])
  cor = df[name].corr(df['Potability'])

  print("Nilai korelasi Pearson dari "+name+" : "+str(cor))
  print("Korelasi antara kedua kolom adalah",end=' ')
  if cor == 0:
      print("no correlation")
  elif cor == 1:
      print("perfect positive correlation")
  elif cor == -1:
      print("perfect negative correlation")
  elif 0 < cor < 1:
      print("positive correlation")
  elif -1 < cor < 0:
      print("negative correlation")
  else:
      print("error (nilai korelasi tidak berada pada -1 dan 1)")
```
